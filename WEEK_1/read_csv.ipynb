```python
import pandas as pd

def read_csv_to_dataframe(file_path):
    """
    Reads a CSV file separated by commas and converts it into a DataFrame.
    
    Parameters:
        file_path (str): Path to the CSV file.
    
    Returns:
        pd.DataFrame: Data from the CSV file as a DataFrame.
    """
    df = pd.read_csv(file_path)
    return df

```

```python
terminos_df = read_csv_to_dataframe("Terminos_lagoon_TA_DIC_2023_RawData.csv")
```
