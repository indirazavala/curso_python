## 09_nonparametric_mann_whitney

### Exercise: "Evaluate the Mann-Whitney U test for DIC and temperatura. Create boxplots and violin plots for these analyses." 

```python
def import_csv(file):
    # Read the file into a DataFrame: df
    import pandas as pd
    return pd.read_csv(file)
```

```python
# Path to the file to be imported
path = "E:\prueba\Terminos_lagoon_TA_DIC_2023_RawData.csv"

# Import the file
co2_data = import_csv(path)
```

```python
from scipy.stats import mannwhitneyu
import seaborn as sns
import matplotlib.pyplot as plt
```

```python
# --- Definimos funci√≥n para aplicar el test ---
def mann_whitney_test(data1, data2, alpha=0.05):
    """
    Performs the Mann-Whitney U test and prints U statistic, p-value, and result.
    """
    result = mannwhitneyu(data1, data2, alternative='two-sided')
    
    print(f"U statistic = {result.statistic}")
    print(f"p-value = {result.pvalue}")

    if result.pvalue < alpha:
        print("The difference is statistically significant.")
    else:
        print("The difference is not statistically significant.")
```

```python
# --- Boxplots y Violinplots para DIC ---
plt.figure(figsize=(8,6))
sns.boxplot(x="season", y="dic_micromol_kg", data=co2_data, palette="pastel")
plt.title("Boxplot of DIC by Season")
plt.xlabel("Season")
plt.ylabel("DIC (¬µmol/kg)")
plt.show()

plt.figure(figsize=(8,6))
sns.violinplot(x="season", y="dic_micromol_kg", data=co2_data, palette="muted")
plt.title("Violin Plot of DIC by Season")
plt.xlabel("Season")
plt.ylabel("DIC (¬µmol/kg)")
plt.show()
```

```python
# --- Mann-Whitney U Test para DIC ---
dic_dry = co2_data.loc[co2_data["season"] == "Dry", "dic_micromol_kg"]
dic_rainy = co2_data.loc[co2_data["season"] == "Rainy", "dic_micromol_kg"]

print("\nMann-Whitney U Test for DIC:")
mann_whitney_test(dic_dry, dic_rainy)
```

```python
# --- Boxplots y Violinplots para Temperatura ---
plt.figure(figsize=(8,6))
sns.boxplot(x="season", y="temp_c", data=co2_data, palette="Set2")
plt.title("Boxplot of Temperature by Season")
plt.xlabel("Season")
plt.ylabel("Temperature (¬∞C)")
plt.show()

plt.figure(figsize=(8,6))
sns.violinplot(x="season", y="temp_c", data=co2_data, palette="coolwarm")
plt.title("Violin Plot of Temperature by Season")
plt.xlabel("Season")
plt.ylabel("Temperature (¬∞C)")
plt.show()
```

```python
# --- Mann-Whitney U Test para Temperatura ---
temp_dry = co2_data.loc[co2_data["season"] == "Dry", "temp_c"]
temp_rainy = co2_data.loc[co2_data["season"] == "Rainy", "temp_c"]

print("\nMann-Whitney U Test for Temperature:")
mann_whitney_test(temp_dry, temp_rainy)
```
