```python
import pandas as pd
import matplotlib.pyplot as plt

# Cargar archivo CSV
df = pd.read_csv(r"E:\prueba\Terminos_lagoon_TA_DIC_2023_RawData.csv")


```

```python
# Histograma de DIC
plt.hist(df["dic_micromol_kg"].dropna(), bins=20, color="blue", edgecolor="black")
plt.title("Histograma de dic_micromol_kg")
plt.xlabel("DIC (µmol/kg)")
plt.ylabel("Frecuencia")
plt.show()
```

```python
# Histograma de TA
plt.hist(df["ta_micromol_kg"].dropna(), bins=20, color="green", edgecolor="black")
plt.title("Histograma de ta_micromol_kg")
plt.xlabel("TA (µmol/kg)")
plt.ylabel("Frecuencia")
plt.show()
```
