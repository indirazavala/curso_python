```python
import pandas as pd
import matplotlib.pyplot as plt
import seaborn as sns

# Configuraci√≥n general de estilo
sns.set_theme(style="whitegrid", palette="muted")

# Cargar base de datos de ejemplo (ajusta el path a tu archivo)
CO2Data = pd.read_csv("E:/prueba/Terminos_lagoon_TA_DIC_2023_RawData.csv")
```

```python
# =========================================================================
# 1) Cambio de unidades en la etiqueta del eje
# =========================================================================
plt.figure(figsize=(8, 6))
sns.histplot(CO2Data["dic_micromol_kg"]/1000, bins=20, kde=True, color="#1f77b4")
plt.title("Distribuci√≥n de DIC", fontsize=14, fontweight="bold")
plt.xlabel("DIC (mmol/kg)", fontsize=12)   # Cambi√© de ¬µmol/kg a mmol/kg en el label
plt.ylabel("Frecuencia", fontsize=12)
plt.show()
```

```python
# =========================================================================
# 2) Boxplot comparando √°reas y temporadas
# =========================================================================
plt.figure(figsize=(10, 6))
sns.boxplot(
    data=CO2Data,
    x="season",
    y="ta_micromol_kg",
    hue="area",
    palette="Set3"
)
plt.title("TA por temporada y √°rea", fontsize=14, fontweight="bold")
plt.xlabel("Temporada", fontsize=12)
plt.ylabel("TA ($\\mu mol \\; kg^{-1}$)", fontsize=12)
plt.legend(title="√Årea", loc="upper right")
plt.show()
```

```python
# =========================================================================
# 3) Scatterplot con regresi√≥n (DIC vs Salinidad por estuario)
# =========================================================================
sns.lmplot(
    data=CO2Data,
    x="sal_psu",
    y="dic_micromol_kg",
    hue="season",
    col="estuary",
    height=5,
    aspect=1,
    palette="coolwarm",
    scatter_kws={"s": 40, "alpha": 0.7},  # tama√±o y transparencia de los puntos
    line_kws={"lw": 2, "alpha": 0.9}      # estilo de la l√≠nea de regresi√≥n
)
plt.subplots_adjust(top=0.85)
plt.suptitle("Relaci√≥n entre DIC y Salinidad por Estuario", fontsize=14, fontweight="bold")
plt.show()
```

```python
import seaborn as sns
import matplotlib.pyplot as plt
import pandas as pd

# Leer la base (ajusta la ruta a la tuya)
CO2Data = pd.read_csv("E:\prueba\Terminos_lagoon_TA_DIC_2023_RawData.csv")

# Estilo
sns.set_theme(style="ticks")

# Crear figura
f, ax = plt.subplots(figsize=(8, 6))

# Boxplot horizontal de DIC agrupado por estuario
sns.boxplot(
    data=CO2Data, x="dic_micromol_kg", y="estuary",
    whis=[0, 100], width=.6, palette="coolwarm"
)

# Agregar puntos individuales
sns.stripplot(
    data=CO2Data, x="dic_micromol_kg", y="estuary",
    size=4, color=".3"
)

# Personalizaci√≥n
ax.set_xlabel("DIC (¬µmol/kg)")
ax.set_ylabel("Estuario")
ax.xaxis.grid(True)
sns.despine(trim=True, left=True)

plt.show()
```
